```
# First we group by lesion id to see how many images per lesion we have
duplicates_df = tile_df.iloc[:,:2].groupby('lesion_id').count()

# Then, we filter for all lesions that only have one associated image. Thus, we can be assured 
# that the same lesion will not be found in the training as well as the test images. 
duplicates_df = duplicates_df[duplicates_df['image_id'] == 1]
duplicates_df.reset_index(inplace=True)
duplicates_df.head()
```

```
# Prepare a function to identify lesions with no duplicate images
def identify_duplicates(x):
    
    unique_list = list(duplicates_df['lesion_id'])
    
    if x in unique_list:
        return 'no_duplicates'
    else:
        return 'has_duplicates'
    
# Add a column to our dataframe that will mark images with duplicate pairs
tile_df['duplicates'] = tile_df['lesion_id']

# Use our function to map lesion ID to a 'duplicate' marker in the main dataframe
tile_df['duplicates'] = tile_df['duplicates'].apply(identify_duplicates)

tile_df.head(3)
```

```
tile_df['duplicates'].value_counts()
```

```
# We will use images with no duplicates to build our test and validation sets
no_duplicates_df = tile_df[tile_df['duplicates'] == 'no_duplicates']
no_duplicates_df.shape
```

```
# Now we select a portion of our 'no_duplicate' images to create our test and validation sets. 
# Note that test_size = .55 so that we end up with ~3k images which represents about 30% of our 
# full dataset of ~10k images. 
y = no_duplicates_df['cell_type_idx']
_, test_val_df = train_test_split(no_duplicates_df, 
                                    test_size = .55, 
                                    stratify=y, 
                                    random_state=108) 
test_val_df.shape
```

```
# Now we choose 20% of our test images to create our validation set 
y = test_val_df['cell_type_idx']
test_df, valid_df = train_test_split(test_val_df, 
                                       test_size = .2, 
                                       stratify=y, 
                                       random_state=108) 
```

```
# Now we will use the remaining images to create our training dataframe

# Define a function to identify if an image is in our test or validation sets. 
def identify_val_rows(x):
    # create a list of all the lesion_id's in the val set
    test_list = list(test_df['image_id'])
    val_list = list(valid_df['image_id'])
    
    if str(x) in test_list or str(x) in val_list:
        return 'test_val'
    else:
        return 'train'

# Create a new colum that is a copy of the image_id column
tile_df['train_or_val'] = tile_df['image_id']

# Apply our function to this new column to create a label for images in test or validation sets
tile_df['train_or_val'] = tile_df['train_or_val'].apply(identify_val_rows)
   
# Finally, we create a training dataframe with the appropriate images
train_df = tile_df[tile_df['train_or_val'] == 'train']
```

```
# First, we create a dataframe with each class name and it's associated index
aug_df = train_df.iloc[:,8:10].drop_duplicates().\
                sort_values('cell_type_idx').reset_index().drop('index', axis=1)

# Now, we define our 'aug_multiplier' in terms of the size of our largest class
target_image_count = len(train_df[train_df['dx'] == 'nv'])  # 4276

aug_df['aug_multiplier'] = [target_image_count//i for i in tile_df['cell_type'].value_counts()]
aug_df
```

```
# Create a dictionary with keys = lesion_index and values = aumentation_multiplier
temp_dict = aug_df.drop(['cell_type', 'cell_type_idx'], axis=1).to_dict()
aug_dict = temp_dict.get('aug_multiplier')
aug_dict
```

```
# Create a new column in our training dataframe that has our augmentation value for each sample
train_df['aug_multiplier'] = train_df['cell_type_idx'].map(lambda x: aug_dict.get(x))
```

```
# Create our balanced directory which will contain our final augmented dataset
balanced_dir = 'balanced_dir'
#os.mkdir(balanced_dir)
os.makedirs(balanced_dir, exist_ok=True)


# Add folders for train, test and validation sets
train_dir = os.path.join(balanced_dir, 'train')
os.makedirs(train_dir, exist_ok=True)
test_dir = os.path.join(balanced_dir, 'test')
os.makedirs(test_dir, exist_ok=True)
valid_dir = os.path.join(balanced_dir, 'valid')
os.makedirs(valid_dir, exist_ok=True)


# Create a folder for each class label
def create_class_directories(path):
    nv = os.path.join(path, 'nv')
    os.makedirs(nv, exist_ok=True)
    mel = os.path.join(path, 'mel')
    os.makedirs(mel, exist_ok=True)
    bkl = os.path.join(path, 'bkl')
    os.makedirs(bkl, exist_ok=True)
    bcc = os.path.join(path, 'bcc')
    os.makedirs(bcc, exist_ok=True)
    akiec = os.path.join(path, 'akiec')
    os.makedirs(akiec, exist_ok=True)
    vasc = os.path.join(path, 'vasc')
    os.makedirs(vasc, exist_ok=True)
    df = os.path.join(path, 'df')
    os.makedirs(df, exist_ok=True)

# Populate each folder 
create_class_directories(os.path.join(train_dir))
create_class_directories(os.path.join(test_dir))
create_class_directories(os.path.join(valid_dir))
```

```
# Initialize the data generator we will use to create our augmented data
datagen = ImageDataGenerator(
        rescale=None,
        rotation_range=30,
        width_shift_range=0.1,
        height_shift_range=0.1,
        shear_range=0.1,
        zoom_range=0.1,
        horizontal_flip=True,
        vertical_flip=True,
        featurewise_center=False,  
        samplewise_center=False,
        featurewise_std_normalization=False,
        samplewise_std_normalization=False,
        fill_mode='reflect')
```

```
# This loop will generate n augmented images per input sample where n is our aug_multiplier
for i, row in train_df.iterrows():
    img = load_img(row['path'])  # this is a PIL image
    x = img_to_array(img)  # this is a Numpy array with shape (600, 450, 3)
    x = x.reshape((1,) + x.shape)  # this is a Numpy array with shape (1, 600, 450, 3)

    counter = 0
    if row['aug_multiplier'] != 0:  # Skip augmentation for class 'nv'
        for batch in datagen.flow(x, batch_size = 1,
                                    save_to_dir = os.path.join(balanced_dir, 'train', row['dx']),
                                    save_format = 'jpeg'):
            counter += 1
            if counter > row['aug_multiplier']:
                break  # Move on to the next image once the correct number has been generated
```

```
# This function will display the number of files that are present in each folder
def display_image_count():
    for folder in os.listdir(balanced_dir):
        if folder == '.ipynb_checkpoints':
            pass
        else:
            print('#'* 60)
            folder_total = 0
            for path_ext in lesion_index_dict.keys():
                folder_total += len(os.listdir(os.path.join(balanced_dir, folder, path_ext)))
                print(f'Directory: "{folder}/{path_ext}" contains\
                      {len(os.listdir(os.path.join(balanced_dir, folder, path_ext)))} images')
            print("\n", " " * 35, f"Total {folder} images: {folder_total}")
```

```
# Define transformations for data augmentation
tfms = get_transforms(do_flip=True,  
                      max_rotate=10,
                      max_zoom=1.1,
                      max_warp=0.2)

# Build dataset by applying transforms to the data from our directory
data = (ImageList.from_folder(balanced_dir)
        .split_by_folder()          
        .label_from_folder()
        .transform(tfms, size=224)
        .databunch()
        .normalize(imagenet_stats))  # Our version of mobilepretrained 
```

```
# The correct slice was taken from [https://github.com/fastai/fastai/pull/2456]
wd=1e-2

mobilenet_split = lambda m: (m[0][0][10], m[1])
arch  = torchvision.models.mobilenet_v2
learn = cnn_learner(data, arch, cut=-1, split_on=mobilenet_split, wd=wd, metrics=[accuracy])
```

```
learn.lr_find();
learn.recorder.plot();
```

```
#Set our learning rate to the value where learning is fastest and loss 
# is still decreasing.
lr=3e-3

# This function uses our input lr as an anchor and sweeps through a range 
# in order to search out the best local minima.
learn.fit_one_cycle(10, slice(lr), pct_start=0.9)
```

```
# Save the current state of our model
learn.save('mobile_v1_stage-1')
```

```
# This function creates a display of the models prediction and confidence levels
def plot_prediction(learner, index):
  data = learner.data.train_ds[index][0]
  pred = learner.predict(data)
  classes = learner.data.classes

  prediction = pd.DataFrame(to_np(pred[2]*100), columns=['Confidence'])
  prediction['Classes'] = classes
  prediction = prediction.sort_values(by='Confidence', ascending=False)

  fig = plt.figure(figsize=(12, 5))
  ax1 = fig.add_subplot(121)
  show_image(data, figsize=(5, 5), ax=ax1)
  ax2 = fig.add_subplot(122)
  sns.set_color_codes("pastel")
  sns.barplot(x='Confidence', y='Classes', data=prediction,
              label="Total", color="b")
  ax2.set_title(f'Actual: {learn.data.train_ds[index][1]}')
```
